## Finding outliers

```python
df['Age_std'] = ((df['Age'] - df['Age'].mean()).apply(abs) / df['Age'].std())
```

```python
df.sort_values(by='Age', ascending=True).head(10)
```

```python
df['Fare_std'] = ((df['Fare'] - df['Fare'].mean()).apply(abs) / df['Fare'].std())
```

```python
df.sort_values(by='Fare', ascending=False).head(60)
```

## Skewness and Kurtosis

```python
df['Age'].skew()
```

```python
df['Fare'].skew()
```

Although we need a frame of reference in order to know if the value is good or not, we know that the closest to 0 the better. Therefore, we can consider skewness to be bell shaped for age but for fare we have a lot of the mass of the distribution concentrated on the left of the figure (with very expensive tickets being outliers). This means we probably shouldn't consider them and take them out of our database (taking around 75 different people who paid the fare at a very high price).

```python
df['Age'].kurt()
```

Platykurtic distribution, which means the distribution has fewer and less extreme outliers than the normal distribution.

```python
df['Fare'].kurt()
```

Leptokurtic distribution, and by a big margin. This is just another way to contue what was said before with the high amount of outliers we have.
