### Exercise 2 -  Calculating Correlation in Pandas


We will be using a dataset "Boston house prices". Calculate the Pearson correlation between "MEDV-Median value of owner-occupied homes in $1000's)" and all the variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets 
data = datasets.load_boston() 
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
data.target
```

```python
target = pd.DataFrame(data.target, columns=["MEDV"]) 
target.head()
```

```python
print (data.DESCR)
```

```python
df2 = df.assign(target=target.values)
df2.head()
```

```python
df2.rename(columns={'target': 'MEDV'}, inplace=True)
df2.head()
```

```python
# pairwise correlation Pearson
df2.corr(method='pearson')
```

```python
df2.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# Compute the correlation matrix

corr = df2.corr(method='pearson')

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# pairwise correlation Spearman
df2.corr(method='spearman')
```

```python
df2.corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# Compute the correlation matrix

corr = df2.corr(method='spearman')

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Note:

- HOW TO INTERPRET A CORRELATION COEFFICIENT R?
    
The value of r is always between +1 and –1. To interpret its value, see which of the following values your correlation r is closest to:
        
Exactly –1. A perfect downhill (negative) linear relationship

–0.70. A strong downhill (negative) linear relationship

–0.50. A moderate downhill (negative) relationship

–0.30. A weak downhill (negative) linear relationship

0. No linear relationship

+0.30. A weak uphill (positive) linear relationship

+0.50. A moderate uphill (positive) relationship

+0.70. A strong uphill (positive) linear relationship

Exactly +1. A perfect uphill (positive) linear relationship
