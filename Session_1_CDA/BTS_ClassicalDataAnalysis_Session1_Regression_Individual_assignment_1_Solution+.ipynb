### 1.1. Regression model with Statsmodels and without a constant:

```python
import statsmodels.api as sm

X = df["sepal width (cm)"]
y = df["sepal length (cm)"]
```

```python
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
```

```python
predictions = model.predict(X) 
```

```python
model.summary()
```

### Interpreting the Table 

The coefficient of 1.8717 means that as the sepal width (cm variable increases by 1, the predicted value of sepal length (cm) increases by 1.8717. We can see here that this model has a much high R-squared value 0.957, meaning that this model explains 95.7% of the variance in our dependent variable. 

### 1.2. Regression model with Statsmodels and with a constant:

```python
import statsmodels.api as sm 
X = df["sepal width (cm)"]
y = df["sepal length (cm)"]
```

```python
X = sm.add_constant(X) 
```

```python
model = sm.OLS(y, X).fit() 
```

```python
predictions = model.predict(X)
```

```python
model.summary()
```

### Interpreting the Table 
With the constant term the coefficients are different. Without a constant we are forcing our model to go through the origin, but now we have a y-intercept at 6.4812. We also changed the slope of the sepal width (cm) predictor from 1.8717 to -0.2089.  A few other important values are modification of the R-squared, from 0.957 to 0.012.
